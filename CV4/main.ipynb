## Imports

```python
import algorithm, visualization, distances
from classes import Config
from IPython.display import HTML
```

## Configuration

```python
# Configuration
config: Config = Config(
    population_size=2000,                   # Number of populations
    tournament_selection_size=4,            # Number of cities in selection for best cities
    mutation_rate=0.1,                      # Chance for mutation
    crossover_rate=0.9,                     # Chance to crossover
    generations=200,                        # Number of generations to create
    target=0,                               # Target minimum distance (optional)
    cities_file="data/TSP51.txt",           # None for random cities or path to data file, ex. "data/TSP51.txt"
    cities_length=30,                       # If cities are generated - this is number of cities
    distance_method=distances.euclidean     # Distance calculation function (for cities)
)
```

## TSP - genetic algorithm

```python
first_population, first_result = algorithm.population_selection(config)
result, progress = algorithm.algorithm_core(first_population, config)

print("\nGenerations: " + str(config.generations))
print("Start distance: " + str(round(first_result[0], 2)))
print("Result distance: " + str(round(result[0], 2)))
```

## Animated progress visualization

```python
# Create progress animation
anim_1 = visualization.visualize_animated_progress(config, progress)
HTML(anim_1.to_html5_video())
```
